## step1 AndroidManifest.xml 데이터 불러와서 파싱하는 함수 선언

```python
import os
import pandas as pd

dir_names = ["Benign_samples_1st_1500", "Malware_samples_1st_500"]

def parse_permission(menifest_name):
    permissions_dict = dict()
    with open(menifest_name) as f :
        for l in f.readlines():
            l = l.strip()
            if l.startswith("<uses-permission android"):
                perm = l.split('"')[1].split(".")[-1]
                permissions_dict[perm] = 1
    
    return permissions_dict

def bring_smali_list_from_dir(hash_value, permission_list, dirname):
    #permission_list = dict()
    menifest_name = dirname + "_apktools/" + hash_value + '/AndroidManifest.xml'
    permissions_dict = parse_permission(menifest_name)
    permission_list[hash_value] = permissions_dict
                
    return permission_list
```

## step3 Malware 악성 파일 csv 파일 불러오기

```python
malware_samples = pd.read_csv('Malware_random_list_1st_500.csv')
malware_samples.index = malware_samples['filename']
del malware_samples['filename']
del malware_samples['fsecure_scan_result']
del malware_samples['scan_date']

family = sorted(list(set('family_' + malware_samples.family)))
print(family)
malware_samples = malware_samples.sort_values('family')
malware_samples
```

## step4 Categorical Data

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
import numpy as np

labelencoder_y = LabelEncoder()
malware_samples.family = labelencoder_y.fit_transform(malware_samples.family)
onehotencoder = OneHotEncoder(categorical_features=[1])
X = onehotencoder.fit_transform(malware_samples).toarray()

#columns = set(malware_samples.family).add('set(malware_samples.family)
family.append('class')
bm_df1 = pd.DataFrame(X, columns = family, index = malware_samples.index )
bm_df1 = pd.concat([bm_df1,benign_sample_list]).fillna(0)
del bm_df1['family']
bm_df1.sort_index(axis=1)
```
